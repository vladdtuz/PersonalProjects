Prices of Hass Avocados 

What insights can we gain from it? Are millennials really hiking up the prices?

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('avocado.csv')
```

```python
data.head()
```

```python
data.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
data.set_index('Date',inplace=True)
```

```python
data.head()
```

```python
sns.lineplot(x=data.index,y='Small Bags',data=data,hue='type')
```

```python
sns.lineplot(x=data.index,y='Large Bags',data=data,hue='type')
```

```python
sns.lineplot(x=data.index,y='XLarge Bags',data=data,hue='type')
```
